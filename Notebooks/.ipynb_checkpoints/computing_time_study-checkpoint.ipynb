# Study of the computing time

```python
%run preamble_notebooks.py
danish = pd.read_csv("../Data/Danish/danish.csv").x
computing_time_free = pd.read_csv("../Data/Danish/computing_time_free.csv", sep = ';' )
computing_time_fixed = pd.read_csv("../Data/Danish/computing_time_fixed.csv",  sep = ';')
computing_time_free["threshold_method"] = "free"
computing_time_fixed["threshold_method"] = "fixed"
computing_time_fixed
```

```python
model_types = ["disjoint", "continuous"]
for model_type in model_types:
    print(model_type)
    df = computing_time_fixed.loc[computing_time_fixed.model_type == model_type]
    print(df.pivot(index = ["body", "tail"], columns=["paralell", "popSize"
                                                     ],values="computing_time").to_latex(float_format="%.0f"))
```
